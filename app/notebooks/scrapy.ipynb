````python
import re
def to_markdown(self, html):
    markdown = markdownify(
        html,
        heading_style="ATX",
        escape_asterisks=False,
        escape_underscores=False,
        escape_misc=False,
        sup_symbol="<sup>",
        sub_symbol="<sub>",
    )
    # Clean up markdown text: consolidate newlines; replace non-breaking spaces
    markdown = re.sub(r"\n\n+", "\n\n", markdown).replace('\u00A0', ' ')
    return markdown.strip()
````

````python
import json
output_json = "edd/edd-about-sdi.json"
with open(output_json, 'r') as fp:
    for data in json.load(fp):
        print("\n\n==============================")
        print(data["title"], data["url"])
        if "main_content" in data:
            print("\n------- @MAIN_CONTENT:\n")
            print(data["main_content"])
        if "main_primary" in data:
            print("\n------- @MAIN_PRIMARY:\n")
            print(data["main_primary"])
        if "nonaccordion" in data:
            print("\n------- @NONACCORDION:")
            print(data["nonaccordion"])
        if "accordions" in data:
            print("\n------- @ACCORDIONS:")
            for heading,paras in data["accordions"].items():
                print(f"\n---- ## {heading}:\n")
                for para in paras:
                    print(para)
    
````
